```python
import pandas as pd
import glob
```

```python
#make a list of all the csvs
filenames_list = glob.glob('data/20*.csv')
filenames_list
```

```python
#read in each csv into a df
dfs_list = [pd.read_csv(filename) for filename in filenames_list]
```

```python
#concactenate all the nyc dfs
all_df = pd.concat(dfs_list, ignore_index=True)
all_df.head()
```

```python
#delete rows where trip duration is greater than 1 day (1440 mins)
all_df = all_df[all_df['tripduration'] < 1441]
all_df.dtypes
```

```python
# overwriting gender with replaced value of 
all_df["gender"]= all_df['gender'].replace(1, 'Male')
all_df["gender"]= all_df['gender'].replace(2, 'Female')
all_df = all_df[all_df['gender'] != 0]
all_df.head()
```

```python
# saving the dataframe 
all_df.to_csv('citi_combined.csv', index=False)
```
